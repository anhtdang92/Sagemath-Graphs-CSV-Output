```python
length = len(chunks)
g6_array = []

for i in range(1, length, 1):
    Gx = Graph(chunks[i])
    g6_array.append(Gx)
    
    
# make a list of int counting from 1 to total length + 1    
num_array = []

for j in range(1, length+1, 1):
    num_array.append(j)
    
num_array.insert(0, '#')
```

```python
graph_order = []
for x in g6_array:
    order = x.order()
    graph_order.append(order)
    
graph_order.insert(0, 'Order')

graph_size = []
for x in g6_array:
    size = x.size()
    graph_size.append(size)
    
graph_size.insert(0, 'Size')
    
graph_deg_seq = []
for x in g6_array:
    deg_seq = x.degree_sequence()
    graph_deg_seq.append(deg_seq)
    
graph_deg_seq.insert(0, 'Degree Sequence')
    
graph_con_com_num = []
for x in g6_array:
    con_com_num = x.connected_components_number()
    graph_con_com_num.append(con_com_num)
    
graph_con_com_num.insert(0, '# Connected Component')

# transpose lists
rows = zip(num_array,chunks,graph_order,graph_size,graph_deg_seq,graph_con_com_num)
```
